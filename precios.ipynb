## Entrenamos una inteligencia artificial

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path_to_data = ('NY_airbnb.csv')
df = pd.read_csv(path_to_data)

df.head()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.drop(["id", "name", "host_id", "host_name", "number_of_reviews", "minimum_nights", "last_review", "reviews_per_month", "calculated_host_listings_count"], axis=1, inplace=True)
```

```python
df.head()
```

```python
s2=df[df.price < 500]
```

```python
X = s2.drop("price", axis=1)
y = s2["price"]
```

```python
X.head()
```

```python
y.head()
```

```python
categoricas = ["neighbourhood_group", "neighbourhood", "room_type"] 
numericas = ["latitude", "longitude","availability_365"]
```

```python
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer

transformer_numerico = ("transformer_numerico", MinMaxScaler(), numericas)
transformer_categorico = ("transformer_categorico", OneHotEncoder(), categoricas)

transformer = ColumnTransformer([transformer_numerico, transformer_categorico], remainder="passthrough")
```

```python
X = transformer.fit_transform(X).toarray()
```

```python
pd.DataFrame(X, columns = transformer.get_feature_names_out())
```

```python
transformer.output_indices_
```

## Entrenamos

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR

from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
def train_evaluate(nombre_modelo, modelo):
    mod = modelo()
    mod.fit(X_train, y_train)
    
    y_predict = mod.predict(X_test)
    
    mae = mean_absolute_error(y_test, y_predict)
    mse = mean_squared_error(y_test, y_predict)
    rmse = mean_squared_error(y_test, y_predict, squared=False)
    r2 = r2_score(y_test,y_predict)
    
    y_pred_train =mod.predict(X_train)
    mae_train = mean_absolute_error(y_train,y_pred_train)     
    mse_train = mean_squared_error(y_train,y_pred_train)     
    rmse_train = mean_squared_error(y_train,y_pred_train)     
    r2_train = r2_score(y_train,y_pred_train)
    
    print(nombre_modelo)
    print()
    print(f"MAE: {mae}")
    print(f"MSE: {mse}")
    print(f"RMSE: {rmse}")
    print(f"R2: {r2}")
    print()
    
    
    print(nombre_modelo)
    print()
    print(f"mae_train: {mae_train}")
    print(f"mse_train: {mse_train}")
    print(f"rmse_train: {rmse_train}")
    print(f"r2_train: {r2_train}")
    print()
          
```

```python
# regresion lineal

train_evaluate("LinearRegresion", LinearRegression)
```

```python
# arbol de decision

train_evaluate("Decission Tree", DecisionTreeRegressor)
```

```python
# regresion vecinos cercanos

train_evaluate("KNeighborsRegressor", KNeighborsRegressor)
```

```python
# randon forest regressor

train_evaluate("RandomForestRegressor", RandomForestRegressor)
```

```python
# support vector machine regresion

train_evaluate("SVR", SVR)
```
